Multiprocessing runs multiple processes, each with its own memory space and typically on separate CPUs.

2. What are the challenges associated with memory management in Python?
ans. Memory management in Python is automatic through reference counting and garbage collection, but it still presents several challenges that can affect performance and stability.​

Key Challenges in Python Memory Management
1. Circular References
Python uses reference counting to track objects. When two or more objects reference each other (a circular reference), their reference counts never drop to zero, preventing automatic deallocation. While the garbage collector can often handle these cases, it may not detect or clean all circular references efficiently, particularly if objects define custom __del__ methods.
2. Memory Leaks
Memory leaks occur when allocated memory is not freed even though it’s no longer needed—often due to lingering references, large unused data structures, or mismanagement in external libraries. These leaks accumulate over time and degrade performance, especially in long-running processes like web servers or data pipelines.​

3. Fragmentation and Retained Heap Memory
Even after objects are freed, Python’s memory allocator (pymalloc) may not return that memory to the operating system immediately. This leads to fragmentation and inflated memory usage, which can become a problem in long-running or memory-intensive applications.

7. :Explain the difference between apply() and map() in Pandas.
Use map() when you want to transform or substitute values in a Series using a simple function, dictionary, or mapping, operating element-wise.

Use apply() for more complex operations across rows or columns in a DataFrame or for element-wise function application on a Series when the function involves aggregation or interaction across elements.​

This distinction allows map() to be a lightweight, fast option for straightforward mappings and apply() to handle more versatile and computationally complex transformations.

```python
#Calculate the correlation matrix using Seaborn and visualize it with a heatmap.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame
df = pd.DataFrame({
    'Math': [90, 85, 80, 75, 70],
    'Science': [85, 80, 75, 70, 65],
    'English': [80, 75, 70, 65, 60]
})

# Calculate the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix with a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Add a title
plt.title('Subject Correlation Heatmap')

# Display the plot
plt.show()
```

```python
#Write a Python program that logs an error message to a log file when a division by zero exception occurs.
import logging


logging.basicConfig(filename='error_log.txt', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError as e:
        logging.error("Division by zero error occurred: %s", e)
        print("Error logged: Division by zero.")


safe_divide(10, 0)
```

```python
#Write a program that handles both IndexError and KeyError using a try-except block.
# Program to handle IndexError and KeyError

def handle_errors():
    my_list = [10, 20, 30]
    my_dict = {"name": "Alice", "age": 25}

    try:
        # Attempt to access invalid list index and dictionary key
        print(my_list[5])       # This will raise IndexError
        print(my_dict["city"])  # This will raise KeyError
    except IndexError as ie:
        print("IndexError caught:", ie)
    except KeyError as ke:
        print("KeyError caught:", ke)

# Example call
handle_errors()
```
